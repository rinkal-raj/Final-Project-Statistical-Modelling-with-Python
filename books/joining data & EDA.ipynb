```python
import pandas as pd
from pathlib import Path
```

```python
city_bike = pd.read_csv('city_bike.csv')
city_bike.head(5)
```

```python
foursquare_api = pd.read_csv('foursquar_api.csv')
foursquare_api.head(5)
```

```python
yelp_api = pd.read_csv('yelp_api.csv')
yelp_api.head(5)
```

```python
df_merge = yelp_api.merge(city_bike, on = 'bike_station')
```

```python
df_merge
```

```python
df_data = df_merge.merge(foursquare_api, on = 'bike_station')
```

```python
df_data
```

```python
df_data.columns
```

```python
# DROP UNRELEVANT COLUMNS
df_data = df_data.drop(['id_x', 'free_bikes_x', 'empty_slots_x', 'latitude_x', 'longitude_x', 'id' ,'latitude_y','longitude_y',
                       'free_bikes_y' , 'empty_slots_y','fsq_id',], axis=1)
```

```python
df_data.head(5)
```

```python
df_data.shape
```

```python
#COUNT DUPLICATE ROW
duplicate_rows_df = df_data[df_data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df_data.count()      # Used to count the number of rows
```

```python
#REMOVE DUPLICATE DATA
df_data = df_data.drop_duplicates()
df_data.head(5)
df_data.shape
```

```python
df_data.count()
```

```python
#COUNT NULL VALUES
print(df_data.isnull().sum())
```

```python
filepath = Path('merge_data.csv')
filepath.parent.mkdir(parents=True, exist_ok=True)

df_data.to_csv(filepath, index=False)
```
