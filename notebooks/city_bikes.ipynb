Send a request to CityBikes for the city of your choice. 

```python
import requests
import json
import pandas as pd
from pathlib import Path
#Define the end point URL for the Chicago city bike network
endpoint = "https://api.citybik.es/v2/networks/divvy"
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
#Define the end point URL for the Chicago city bike network
endpoint = "https://api.citybik.es/v2/networks/divvy"
# define the query parameter to include specific fields
params = {"fields":"location.latitude,location.longitude,stations.free_bikes,stations.bike_stations"}
#send https requests to url
response = requests.get(endpoint)
#get json data  responce from request
json_data = response.json()
#extract bike details from json format
bike_details = json_data["network"]["stations"]
#print json data output
print(json.dumps(json_data, indent = 4))
```

Put your parsed results into a DataFrame.

```python
Chicago_data = pd.DataFrame(bike_details, columns = ['id','latitude','longitude','free_bikes','empty_slots'])
```

```python
Chicago_data['bike_station'] = 0
Chicago_data.loc[:,'bike_station'] = Chicago_data.index
```
