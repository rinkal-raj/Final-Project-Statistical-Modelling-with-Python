```python
api_key = os.environ["YELP_API_KEY"]
headers = {'Authorization': 'Bearer %s' % api_key}
```

```python
#Define the end point URL for the milano city bike network
endpoint = 'https://api.yelp.com/v3/businesses/search'
```

```python
city_bike = pd.read_csv('city_bike.csv')
city_bike.head(5)
```

```python
N = city_bike.shape[0] 
#N = 10
df_all = pd.DataFrame() 
for bike_station_index in range(0,N):
    lat = city_bike['latitude'].loc[bike_station_index]
    log = city_bike['longitude'].loc[bike_station_index]
    params = {'term':'Restaurants','latitude': lat, 'longitude': log, 'radius': '1000' , 'sort_by' : 'distance' ,'sort_by' : 'rating'} 
   
    req=requests.get(endpoint, params=params, headers=headers)
    # printing the text from the response 
    restuarnt_details = json.loads(req.text)
    # Use json_normalize() to convert JSON to DataFrame

    df2 = pd.json_normalize(restuarnt_details['businesses']) 
    df2['bike_station'] = bike_station_index
    df_all = pd.concat([df_all,df2], ignore_index = True)
    print(bike_station_index)
    #print(df_all['id','bike_station'])
```

```python
df_merge = df_all.merge(city_bike,on = 'bike_station')
```

```python
df_merge.head(20)
```

```python
df_model = df_merge[['id_x','review_count','rating','free_bikes','bike_station','empty_slots','latitude','longitude']]
```
